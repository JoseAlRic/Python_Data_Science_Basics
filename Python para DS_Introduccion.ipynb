**EJEMPLO 10:**

Supongamos que tienes una entrada de usuario que es una cadena de texto que representa un número, y deseas convertirla a un número entero para realizar una operación matemática. Aquí tienes un ejemplo:

```python
entrada_usuario = input("Ingrese un número entero: ")
try:
    numero_entero = int(entrada_usuario)
    resultado = numero_entero * 2
    print(f"El doble del número es: {resultado}")
except ValueError:
    print("La entrada no es un número entero válido.")
```

**int(entrada_usuario)** *se utiliza para convertir la entrada del usuario en un número entero.*

Si el usuario proporciona una entrada válida, se calcula el doble del número y se muestra el resultado. Si la entrada no es un número válido (por ejemplo, si el usuario ingresa una letra en lugar de un número), se maneja la excepción **ValueError** para informar al usuario que la entrada no es válida.

**EJEMPLO 11:**

```python
numero_entero = 42
numero_texto = str(numero_entero)  # Convertir el número entero a cadena
numero_nuevamente = int(numero_texto)  # Convertir la cadena nuevamente a número entero

print(type(numero_entero), numero_entero)
print(type(numero_texto), numero_texto)
print(type(numero_nuevamente), numero_nuevamente )
```

En este ejemplo, comenzamos con un número entero **42**. Luego, convertimos este número en una cadena de texto utilizando **str(numero_entero)**. Después, convertimos la cadena de texto nuevamente a un número entero utilizando **int(numero_texto).**

El resultado final es que **numero_nuevamente** contiene el número entero **42**, pero hemos realizado dos conversiones redundantes entre tipo de datos: primero de entero a cadena y luego de cadena a entero. En la mayoría de los casos, no tendría sentido realizar estas conversiones consecutivas, ya que podríamos trabajar directamente con el número entero original sin necesidad de convertirlo a cadena y luego volver a convertirlo en entero.

Tambien usamos un funcion integrada de python llamada **type()**, esta funcion nos ayuda a dereminar el tipo de dato que contiene nuestra variable

#**5. Funciones**

Una función es un bloque de código reutilizable que realiza una tarea específica. Las funciones permiten organizar el código en piezas más pequeñas y manejables, lo que facilita su comprensión y mantenimiento. Una función puede recibir argumentos (valores de entrada), realizar operaciones y devolver un resultado (valor de salida) opcional.

**EJEMPLO 1:**

```python
def suma(a, b):
    resultado = a + b
    return resultado
```

En este ejemplo:



*  **def** es la palabra clave que se utiliza para definir una función en Python.
*   **suma** es el nombre de la función que hemos elegido.
*   **(a, b)** son los parámetros de entrada que la función que recibira. Estos son los valores que pasamos a la función cuando la llamamos.
*   **resultado = a + b** es la operación que realiza la función, en este caso, suma los valores de a y b y los almacena en la variable resultado.
*  **return resultado** indica que la función devolverá el valor almacenado en la variable
*  **resultado** como resultado de la operación.

Ahora puedes llamar a esta función con diferentes valores:





```python
resultado1 = suma(5, 3)  # Llama a la función con a=5 y b=3
print(resultado1)  # Imprime el resultado, que será 8

resultado2 = suma(10, 7)  # Llama a la función con a=10 y b=7
print(resultado2)  # Imprime el resultado, que será 17
```

**5.1 Funciones Integrada**


La palabra reservada **print** en Python es una función integrada en el lenguaje. Las funciones integradas, también conocidas como funciones incorporadas o funciones nativas, son aquellas que están disponibles en Python sin necesidad de importar módulos adicionales y son parte del núcleo del lenguaje. La función **print** se utiliza para mostrar información en la pantalla, lo que facilita la depuración y la interacción con el usuario.

Diferencias entre una función integrada y una función definida por el usuario:

**Función Integrada (Ejemplo: print):**

**Disponibilidad inmediata:** Las funciones integradas, como **print**, están disponibles en Python sin necesidad de importar bibliotecas o definirlas previamente. Puedes usarlas en cualquier momento.

**Amplia utilidad:** Las funciones integradas son herramientas fundamentales y comunes utilizadas en una amplia variedad de situaciones. **print** se utiliza para mostrar resultados en la consola.

**Eficiencia:** Las funciones integradas suelen estar altamente optimizadas y son eficientes en términos de rendimiento.

Una de las funciones integradas más comunes en Python es **len()**, que se utiliza para obtener la longitud (número de elementos) de una secuencia, como una lista o una cadena. Por ejemplo:

**EJEMPLO 2**

```python
mi_lista = [1, 2, 3, 4, 5]
longitud = len(mi_lista)
print(longitud)  # Esto imprimirá 5, que es la longitud de la lista
```

En este caso, **len()** es una función integrada que nos ayuda a determinar la longitud de la lista **mi_lista**.

**5.2. Funciones definidas**

**Definición personalizada:** Las funciones definidas por el usuario son creadas por el programador para realizar tareas específicas que no están disponibles como funciones integradas.

**Requiere definición previa:** Debes definir una función antes de poder usarla en tu código. Esto implica escribir el código para la función y, en algunos casos, especificar los parámetros de entrada y la lógica de la función.

**Modularidad y reutilización:** Las funciones definidas por el usuario promueven la modularidad y la reutilización del código, ya que puedes crear funciones personalizadas para realizar tareas específicas en múltiples partes de tu programa.

**EJEMPLO 3:**

```python
def presentar(nombre, edad):
    print(f'El nombre es {nombre} y su edad es {edad} años')

# Llamada a la función
presentar('Jose', 25)
```

**Explicación:**

1. **def presentar(nombre, edad):** es la declaración que define la función presentar. Esta función toma dos parámetros de entrada, nombre y edad.

2. **print(f'El nombre es {nombre} y su edad es {edad} años')** es el cuerpo de la función. Aquí, en lugar de usar **return**, simplemente se utiliza **print()** para mostrar un mensaje en la consola. La cadena de formato (marcada por **f'...')** contiene marcadores de posición **{nombre} **y **{edad}** que se llenan con los valores de **nombre** y **edad** respectivamente.

3. La función **presentar** no devuelve un valor específico, lo que significa que se considera que devuelve **None** de forma implícita. Recuerda que **None** es un valor especial en Python que se utiliza para representar la ausencia de un valor.

4. Finalmente, se llama a la función **presentar('Jose', 25)** pasando los valores **'Jose'** y **25** como argumentos. Esto ejecutará el código dentro de la función y mostrará el mensaje en la consola.

**5.3 Funciones con parametro**

Las funciones con parámetros en Python son funciones que aceptan valores de entrada, llamados parámetros, que se utilizan dentro del cuerpo de la función para realizar operaciones o cálculos. Los parámetros permiten que una función sea más flexible y reutilizable, ya que pueden recibir diferentes valores cada vez que se llama la función. Los parámetros se definen entre paréntesis en la declaración de la función y se utilizan como variables locales dentro de la función para trabajar con los valores proporcionados cuando se llama la función.

```python
def saludar(nombre):
    print(f'Hola, {nombre}.')

# Llamada a la función con un argumento
saludar('Juan')
```

**Explicación:**

1. **def saludar(nombre):** es la declaración que define una función llamada **saludar** con un parámetro **nombre**. Este parámetro **nombre** permite que la función acepte un valor de entrada que se utilizará dentro del cuerpo de la función.

2. **print(f'Hola, {nombre}.')** es el cuerpo de la función. Utiliza el valor del parámetro **nombre** para imprimir un mensaje de saludo personalizado en la consola.

3. Luego, la función **saludar** se llama pasando **'Juan'** como argumento. Cuando se llama a la función, el valor 'Juan' se asigna al parámetro **nombre**, y la función imprime "Hola, Juan." en la consola.

En este ejemplo, la función saludar toma un parámetro nombre, lo utiliza para personalizar un mensaje de saludo y luego se llama a la función pasando un nombre específico como argumento. Los parámetros permiten que la función sea más flexible, ya que puedes proporcionar diferentes valores de nombre cada vez que la llames, lo que resulta en saludos personalizados.

**5.4 Funciones sin parametro**

Una función sin parámetros en Python es una función que no recibe ningún valor de entrada o argumento cuando se llama. En otras palabras, no tiene variables que se pasen desde fuera de la función. En su lugar, realiza acciones o cálculos basados únicamente en su propio código interno y no depende de información externa.

**EJEMPLO 4**

```python
def saludar():
  nombre = input('Digite su nombre: ')
  print(f'Hola {nombre} sea bienvenid@ !!!')
```

**Explicación:**

1. **def saludar():** es la declaración que define la función **saludar**. Esta función no toma argumentos, ya que no hay paréntesis con nombres de parámetros en la declaración. Esto significa que la función no espera recibir ningún valor cuando se llama.

2. Dentro del cuerpo de la función, se utiliza **input('Digite su nombre: ')** para solicitar al usuario que ingrese su nombre. La función **input()** permite la entrada de datos desde el teclado y devuelve la entrada como una cadena de texto.

3. Luego, se utiliza **print(f'Hola {nombre} sea bienvenid@ !!!')** para imprimir un mensaje de bienvenida que incluye el nombre ingresado por el usuario. La cadena de formato **(f'...')** se utiliza para insertar el valor de la variable nombre dentro del mensaje.

4. La función **saludar** no tiene una declaración **return**, lo que significa que no devuelve ningún valor específico. En este caso, la función simplemente muestra un mensaje de bienvenida en la consola.

Puedes llamar a esta función **saludar** para que el usuario ingrese su nombre y luego muestre un mensaje de bienvenida personalizado. Por ejemplo:

```python
saludar()
```

Al ejecutar este código, la función saludar solicitará al usuario que ingrese su nombre, y luego imprimirá un mensaje de bienvenida personalizado en función del nombre ingresado.

Las funciones sin parámetros son útiles cuando deseas encapsular una acción o cálculo específico que no depende de datos externos. Son simples de usar y pueden ser reutilizadas en diferentes partes de tu programa cuando necesites realizar la misma tarea sin tener que proporcionar datos adicionales cada vez que se llame la función.

#**6. Lenguajes**

**6.1 Alto Nivel**

Un lenguaje de programación de alto nivel es un lenguaje de programación que se ha diseñado para ser fácil de leer y escribir para los programadores, en lugar de estar diseñado para ser fácilmente comprendido por una computadora. Aquí hay algunas características clave de los lenguajes de alto nivel, con un enfoque en Python:

1. **Legibilidad y sintaxis clara: **Python se destaca por su sintaxis clara y legible. El código Python se asemeja al lenguaje humano y utiliza una estructura que hace que sea más fácil de entender y depurar.

2. **Abstracción de bajo nivel:** Los lenguajes de alto nivel ocultan gran parte de los detalles de bajo nivel de la computadora, como la gestión de memoria y la manipulación de registros de hardware. Esto permite a los programadores centrarse en resolver problemas y no en detalles técnicos.

3. **Portabilidad:**Los programas escritos en lenguajes de alto nivel son más portables, lo que significa que pueden ejecutarse en diferentes sistemas operativos y plataformas sin modificaciones importantes.

4. **Bibliotecas y módulos:** Los lenguajes de alto nivel, como Python, ofrecen una amplia gama de bibliotecas y módulos que simplifican tareas comunes, como el procesamiento de datos, la creación de interfaces gráficas de usuario y la programación web.

5. **Interpretación o compilación:** Python es un lenguaje interpretado, lo que significa que el código fuente se ejecuta línea por línea por un intérprete. Esto permite una mayor flexibilidad y facilidad en el desarrollo, aunque a menudo puede ser más lento que los lenguajes compilados.

**6.2 Bajo Nivel**

Los lenguajes de bajo nivel, como el Lenguaje Assembly, están más cerca de la máquina y ofrecen un mayor control sobre los recursos de hardware, pero también son más difíciles de programar y menos portables. Aquí hay algunas diferencias clave:

1. **Abstracción:** Los lenguajes de bajo nivel tienen una abstracción mínima y requieren que los programadores gestionen directamente los registros de la CPU y la memoria.

2. **Eficiencia:** Los lenguajes de bajo nivel a menudo son más eficientes en términos de uso de recursos de hardware, pero requieren más tiempo y esfuerzo para escribir código.

3. **Portabilidad:** Los programas escritos en lenguajes de bajo nivel son menos portables y a menudo deben adaptarse específicamente para diferentes arquitecturas de CPU.

4. **Legibilidad:** Los lenguajes de bajo nivel son menos legibles y requieren una comprensión profunda de la arquitectura de la computadora.

**7.1 ¿Que es una condicional?**

Los condicionales en Python son estructuras de control que te permiten tomar decisiones en tu programa en función de una condición o expresión booleana. Puedes usar condicionales para ejecutar un bloque de código si una condición es verdadera y, opcionalmente, otro bloque si la condición es falsa. Aquí tienes una breve explicación junto con un ejemplo:



**EJEMPLO 1:**

Supongamos que deseas verificar si una persona es mayor de edad (mayor o igual a 18 años) antes de permitirle ingresar a un lugar. Puedes usar una estructura condicional para hacer esto. Aquí está un ejemplo:

```python
edad = 20  # Edad de la persona

if edad >= 18:
    print("Puede ingresar, es mayor de edad.")
else:
    print("Lo siento, es menor de edad y no puede ingresar.")
```

**Explicación:**

1. **if edad >= 18:** Esta línea inicia una estructura condicional. Verifica si la variable **edad** es mayor o igual a 18.

2. **print("Puede ingresar, es mayor de edad."):** Si la condición del **if** es verdadera (es decir, si la persona es mayor de edad), se ejecutará este bloque de código y se imprimirá "Puede ingresar, es mayor de edad." en la consola.

3. **else**: Esta parte se ejecuta si la condición del **if** es falsa (es decir, si la persona no es mayor de edad).

4. **print("Lo siento, es menor de edad y no puede ingresar."):** Si la condición es falsa, se ejecutará este bloque de código y se imprimirá "Lo siento, es menor de edad y no puede ingresar." en la consola.

Las condicionales en Python te permiten tomar decisiones basadas en condiciones. Puedes ejecutar diferentes bloques de código según si una condición es verdadera o falsa. Esto es fundamental para controlar el flujo de tu programa y tomar decisiones dinámicas en función de los datos o eventos que se produzcan durante la ejecución del programa.

**EJEMPLO 2:**

```python
def verficar_si_puede_conducir(edad):
  if edad >= 18:
    print("Puedes conducir, es mayor de edad.")
  else:
    print("Lo siento, es menor de edad y no puede conducir.")

verficar_si_puede_conducir(12)
```

#**8. Bucles o Loops**

**8.1 ¿Que es un bucle o loop?**


Un bucle, también conocido como loop en inglés, es una estructura de control en programación que permite ejecutar un conjunto de instrucciones repetidamente mientras se cumple una condición o durante un número específico de veces. Los bucles son fundamentales para automatizar tareas repetitivas en un programa. En Python, hay dos tipos principales de bucles: el bucle **for** y el bucle **while**.


**8.2 Bucle for**

1. El bucle **for** se utiliza para iterar sobre una secuencia (como una lista, una tupla, una cadena de texto, etc.) o cualquier objeto iterable en Python.

2. Se ejecuta un número predefinido de veces, una vez por cada elemento en la secuencia.

3. El bucle **for** es especialmente útil cuando se conoce la cantidad de iteraciones necesarias.

```python
nombres = ["Juan", "Ana", "María"]
for nombre in nombres:
    print(nombre)
```

**8.3 Bucle while**

1. El bucle **while** se utiliza para repetir un bloque de código mientras una condición sea verdadera.

2. Puede ejecutarse un número indefinido de veces, y la condición se verifica antes de cada iteración.

3. El bucle **while** es útil cuando no sabes de antemano cuántas veces se ejecutará el bucle.

```python
contador = 0
while contador < 5:
    print(f"Contador: {contador}")
    contador += 1
```

**EJEMPLO 1:**

 **¿Cuál es el objetivo?**

 Para cada una de estas edades queremos saber si está hábil, apto para conducir y que nos haga esa impresión.

```python
#Bucle FUERA de la funcion
edades = [18, 15, 12, 45, 50]

def verificar_si_puede_conducir(edad):
  if edad >= 18:
    print('Usted tiene edad suficiente para conducir')
  else:
    print('Usted NO tiene edad suficiente para conducir')

#para cada edad dentrro de la lista edades:
for edad in edades:
  verificar_si_puede_conducir(edad)
```

**Explicación del bucle for:**


1. **for edad in edades:**: Esta línea inicia un bucle **for** que recorrerá cada elemento de la lista **edades**. En cada iteración, la variable **edad** tomará el valor del próximo elemento de la lista.

2. **verificar_si_puede_conducir(edad):** Dentro del bucle **for**, llamamos a la función **verificar_si_puede_conducir()** y pasamos el valor de **edad** como argumento. Esto significa que la función se ejecutará una vez por cada elemento de la lista **edades**, y la edad correspondiente se verificará en cada llamada.

3. La función **verificar_si_puede_conducir()** verifica si la edad es mayor o igual a 18 y muestra un mensaje apropiado en función de esa condición.

**EJEMPLO 2:**

```python
#Bucle DENTRO de la funcion
edades = [18, 15, 12, 45, 50]

def verificar_si_puede_conducir_con_bucle(edades):
#para cada edad dentrro de la lista edades:
  for edad in edades:
    if edad >= 18:
     print('Usted tiene edad suficiente para conducir')
    else:
      print('Usted NO tiene edad suficiente para conducir')

verificar_si_puede_conducir_con_bucle(edades)
```

**Explicación del bucle for:**

1. La función **verificar_si_puede_conducir_con_bucle()** toma la lista de **edades**como argumento.

2. Dentro de la función, se utiliza un bucle **for** para iterar a través de cada **edad** en la lista **edades**.

3. Luego, se verifica si cada **edad** es mayor o igual a 18. Si es así, se muestra el mensaje "Usted tiene edad suficiente para conducir". Si no, se muestra el mensaje "Usted NO tiene edad suficiente para conducir".

4. Finalmente, se llama a la función **verificar_si_puede_conducir_con_bucle()** pasando la lista de **edades** como argumento.

**8.4 Bucles anidados**

Un bucle anidado es un bucle que se encuentra dentro de otro bucle. En este caso, hay dos bucles **for** anidados. El primer bucle **for** recorre cada **edad** en la lista **edades**, y el segundo bucle **for** se encuentra dentro del primero y recorre la lista **verificaciones**. Esto significa que, para cada **edad** en la lista **edades**, el segundo bucle **for** itera sobre todas las verificaciones previamente almacenadas en la lista **verificaciones**

```python
verificaciones = []
edades = [18, 15, 12]

def verificar_si_puede_conducir_bool(verificaciones,edades):
  for edad in edades:
    if edad >= 18:
     verificaciones.append(True)
    else:
      verificaciones.append(False)
    for verificacion in verificaciones:
      if verificacion == True:
       print('Usted tiene edad suficiente para conducir')
      else:
        print('Usted NO tiene edad suficiente para conducir')

verificar_si_puede_conducir_bool(verificaciones,edades)
```

La función **append()** es una función integrada de Python que se utiliza para agregar un elemento al final de una lista. En este caso, se utiliza para agregar los resultados de la verificación (True o False) a la lista **verificaciones**.

Por ejemplo, en el código **verificaciones.append(True)** agrega el valor **True** al final de la lista verificaciones, y **verificaciones.append(False)** agrega el valor **False**.

1. El código comienza con dos listas: **verificaciones**, que está inicialmente vacía, y **edades**, que contiene edades.

2. Luego, se llama a la función **verificar_si_puede_conducir_bool()** pasando las listas **verificaciones** y **edades** como argumentos.

3. Dentro de la función, el primer bucle **for** recorre cada **edad** en la lista **edades**.

4. En cada iteración, se verifica si la **edad** es mayor o igual a 18. Si es cierto, se agrega **True** a la lista **verificaciones**. Si no, se agrega **False**.

5. Luego, el segundo bucle **for anidado** itera sobre todas las **verificaciones** almacenadas en la lista verificaciones y verifica si alguna es igual a **True**. Si encuentra al menos una verificación **True**, se muestra el mensaje "Usted tiene edad suficiente para conducir". Si todas las verificaciones son **False**, se muestra el mensaje "Usted NO tiene edad suficiente para conducir".

#**9. Bibliotecas**

**9.1 ¿Que es una Biblioteca?**

En Python, una biblioteca (también conocida como módulo o paquete) es un conjunto de funciones, clases y herramientas predefinidas que extienden la funcionalidad del lenguaje base. Estas bibliotecas proporcionan un conjunto de herramientas especializadas y eficientes para tareas específicas, lo que facilita la programación y el análisis de datos en Python. En el campo de la ciencia de datos, hay varias bibliotecas ampliamente utilizadas que son esenciales para realizar tareas de análisis, manipulación y visualización de datos

**9.2. NumPy**

1. **Explicación:** NumPy es la biblioteca fundamental para la computación científica en Python. Proporciona arreglos multidimensionales (arrays) que son eficientes y rápidos para realizar operaciones matemáticas y numéricas en datos.

2. **Uso común:** Manipulación de datos numéricos, cálculos matemáticos, álgebra lineal.

**9.3. Pandas**

1. **Explicación:** pandas es una biblioteca diseñada para la manipulación y el análisis de datos estructurados. Ofrece estructuras de datos flexibles como DataFrames y Series, que permiten trabajar con datos tabulares de manera eficiente.

2. **Uso común:** Análisis exploratorio de datos, limpieza y transformación de datos, manipulación de datos tabulares.

**9.4 Matplotlib**

1. **Explicación:** matplotlib es una biblioteca de visualización de datos que permite crear gráficos de alta calidad en 2D y 3D. Ofrece una amplia variedad de tipos de gráficos y opciones de personalización.

2. **Uso común:** Creación de gráficos y visualizaciones para explorar y comunicar resultados.

**9.5. Seaborn**

1. **Explicación:** Seaborn es una biblioteca de visualización de datos basada en matplotlib que simplifica la creación de gráficos estadísticos atractivos y descriptivos. Facilita la creación de visualizaciones complejas con menos código.

2. **Uso común:** Visualización de relaciones y distribuciones en datos.

**9.6. Scikit-learn**

1. **Explicación:** scikit-learn es una biblioteca de aprendizaje automático (machine learning) que proporciona herramientas y algoritmos para tareas de clasificación, regresión, clustering y más. Es ideal para aplicaciones de aprendizaje automático y minería de datos.

2. **Uso común:** Construcción, entrenamiento y evaluación de modelos de aprendizaje automático.

**9.7. TensorFlow y PyTorch**

1. **Explicación:** TensorFlow y PyTorch son bibliotecas de código abierto para aprendizaje profundo (deep learning). Son ampliamente utilizadas en el desarrollo de redes neuronales y aplicaciones de inteligencia artificial.

2. **Uso común:** Desarrollo y entrenamiento de modelos de aprendizaje profundo.

**9.8. ¿Como se importa una biblioteca?**

Para importar una biblioteca en Python, puedes utilizar la declaración **import** seguida del nombre de la biblioteca que deseas utilizar.

```python
import nombre_de_la_biblioteca
```

Donde **nombre_de_la_biblioteca** es el nombre de la biblioteca que deseas importar. Por ejemplo, si deseas importar la biblioteca **NumPy**, puedes hacerlo de la siguiente manera:

```python
import numpy
```

También es común utilizar un alias o nombre corto para la biblioteca para facilitar su uso en el código. Por ejemplo, puedes importar NumPy con el alias **np** de esta manera:

```python
import numpy as np
```

Ahora, puedes usar **np** para acceder a las funciones y objetos de NumPy en lugar de escribir el nombre completo de la biblioteca. Por ejemplo:

```python
import numpy as np

# Crear un arreglo NumPy
arr = np.array([1, 2, 3])

# Usar una función de NumPy
media = np.mean(arr)
```

**9.9. Importando un modulo de una biblioteca**

Si deseas importar solo un módulo específico de una biblioteca en lugar de la biblioteca completa, puedes utilizar la declaración **from** seguida del nombre de la biblioteca, seguida de la palabra clave **import** y el nombre del módulo que deseas importar. Aquí tienes cómo se hace:

```python
from nombre_de_la_biblioteca import nombre_del_modulo
```

**Donde:**

1. **nombre_de_la_biblioteca** es el nombre de la biblioteca que contiene el módulo que deseas importar.

2. **nombre_del_modulo** es el nombre del módulo específico que deseas importar.
